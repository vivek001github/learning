```python
def fact(n):
    if n == 1 :
        return 1
    else:
        return n*fact(n-1)
    print(fact(4))
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
def lyrics_to_frequencies(lyrics):
   myDict = {}
   for word in lyrics: 
        if word in myDict:  
            myDict[word] += 1
        else:
            myDict[word] = 1
        return myDict
```
