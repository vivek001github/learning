```python
def is_even(i):
    '''
    input: i, a positive int
    returns True if i is even, otherwise False
    '''
    remainder = i % 2  # Added the equals sign for assignment
    return remainder == 0

print("all numbers between 0 and 20: even or not")
for i in range(21):  # Changed the range to include 20 as well
    if is_even(i):
        print(i, "even")
    else:
        print(i, "odd")
```
